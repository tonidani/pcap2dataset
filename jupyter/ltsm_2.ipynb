```python
import pandas as pd, numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, roc_auc_score
from sklearn.utils import class_weight
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
import joblib

def load_and_prepare_data(path, scaler=None, seq_len=5, fit_scaler=False):
    df = pd.read_csv(path)
    df.columns = df.columns.str.strip()
    df["Label"] = df["Label"].astype(str).str.strip()
    df["BinaryLabel"] = (df["Label"].str.upper() != "BENIGN").astype(int)
    df["Timestamp"] = pd.to_datetime(df["Timestamp"])
    df = df.sort_values(["Flow ID", "Timestamp"])

    exclude = ["Flow ID", "Src IP", "Dst IP", "Src Port", "Dst Port", "Protocol", "Timestamp", "Label", "BinaryLabel"]
    num_cols = df.select_dtypes(include=["float", "int"]).columns.difference(exclude)

    for c in num_cols:
        df[c] = pd.to_numeric(df[c], errors="coerce")
        df[c].replace([np.inf, -np.inf], np.nan, inplace=True)
        df[c].fillna(df[c].median(), inplace=True)

    if fit_scaler:
        scaler = StandardScaler()
        X_all = scaler.fit_transform(df[num_cols])
    else:
        X_all = scaler.transform(df[num_cols])

    y_all = df["BinaryLabel"].values

    X_seq, y_seq = [], []
    for _, group in df.groupby("Flow ID"):
        idxs = group.index
        if len(idxs) < seq_len:
            continue
        gx = X_all[idxs]
        gy = y_all[idxs]
        for i in range(len(gx) - seq_len + 1):
            X_seq.append(gx[i:i + seq_len])
            y_seq.append(gy[i + seq_len - 1])

    return np.array(X_seq, dtype="float32"), np.array(y_seq, dtype="int32"), scaler

def build_model(input_shape):
    model = Sequential([
        LSTM(64, input_shape=input_shape, return_sequences=False),
        Dropout(0.3),
        Dense(32, activation='relu'),
        Dropout(0.2),
        Dense(1, activation='sigmoid')
    ])
    model.compile(loss="binary_crossentropy", optimizer=Adam(1e-3), metrics=["accuracy"])
    return model

def evaluate_model(model, X_test, y_test, label="Model"):
    y_pred_prob = model.predict(X_test).flatten()
    y_pred = (y_pred_prob > 0.5).astype(int)
    
    print(f"\n=== Classification Report ({label}) ===")
    print(classification_report(y_test, y_pred))

    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(5, 4))
    plt.imshow(cm, cmap='Blues')
    plt.title(f"Confusion Matrix ({label})")
    plt.colorbar()
    plt.xticks([0, 1], ["Benign", "Attack"])
    plt.yticks([0, 1], ["Benign", "Attack"])
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    for (i, j), val in np.ndenumerate(cm):
        plt.text(j, i, val, ha='center', va='center')
    plt.tight_layout()
    plt.show()

    fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
    auc = roc_auc_score(y_test, y_pred_prob)

    plt.figure()
    plt.plot(fpr, tpr, label=f"{label} (AUC = {auc:.4f})")
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"ROC Curve - {label}")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

def plot_history(hist, label):
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    plt.plot(hist.history["loss"], label="train")
    plt.plot(hist.history["val_loss"], label="val")
    plt.title(f"Loss ({label})")
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(hist.history["accuracy"], label="train")
    plt.plot(hist.history["val_accuracy"], label="val")
    plt.title(f"Accuracy ({label})")
    plt.legend()

    plt.tight_layout()
    plt.show()

# === Dane treningowe
X_all, y_all, scaler = load_and_prepare_data("../work/2025-04-16_01-02-21.csv", fit_scaler=True)
X_train, X_val, y_train, y_val = train_test_split(X_all, y_all, test_size=0.2, stratify=y_all, random_state=42)

# === Model 1 – bazowy (bez class_weight)
model_plain = build_model((X_train.shape[1], X_train.shape[2]))
history_plain = model_plain.fit(X_train, y_train, epochs=30, batch_size=64, validation_data=(X_val, y_val))
plot_history(history_plain, "Model Bez class_weight")

# === Model 2 – z class_weight
cw = class_weight.compute_class_weight("balanced", classes=np.unique(y_train), y=y_train)
class_weights = dict(enumerate(cw))

model_weighted = build_model((X_train.shape[1], X_train.shape[2]))
history_weighted = model_weighted.fit(X_train, y_train, epochs=30, batch_size=64,
                                      validation_data=(X_val, y_val), class_weight=class_weights)
plot_history(history_weighted, "Model Z class_weight")

# Zapisz modele
model_plain.save("lstm_plain.h5")
model_weighted.save("lstm_weighted.h5")
joblib.dump(scaler, "user_lstm_scaler.pkl")

# === Test na nowych danych
X_test, y_test, _ = load_and_prepare_data("../work/2025-04-19_01-02-32.csv", scaler=scaler)

# Ewaluacja
evaluate_model(model_plain, X_test, y_test, "Bez class_weight")
evaluate_model(model_weighted, X_test, y_test, "Z class_weight")
```
